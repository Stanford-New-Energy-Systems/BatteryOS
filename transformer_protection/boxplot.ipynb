```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
path1 = "{}/Code/Results/case_1_9_homes_l2/f1/".format(os.getcwd())
path3 = "{}/Code/Results/case_3_9_homes_l2/f1/".format(os.getcwd())
path4 = "{}/Code/Results/case_4_9_homes_l2/f1/".format(os.getcwd())
path5 = "{}/Code/Results/case_4_9_homes_l2/f3/".format(os.getcwd())
```

```python
type = "determ"

if type == "determ":
    label = "PF"
else:
    label = "MPC"
```

```python
arr1 = np.array([])
arr3 = np.array([])
arr4 = np.array([])
arr5 = np.array([])

for trial in range(1, 11):
    arr1 = np.append(arr1, np.loadtxt(path1 + "cost_list_{}_1_ev_forecast_0_trial_{}.csv".format(type, trial)))
    arr3 = np.append(arr3, np.loadtxt(path3 + "cost_list_{}_3_ev_forecast_0_trial_{}.csv".format(type, trial)))
    arr4 = np.append(arr4, np.loadtxt(path4 + "cost_list_{}_4_ev_forecast_0_trial_{}.csv".format(type, trial)))
    arr5 = np.append(arr5, np.loadtxt(path5 + "cost_list_{}_4_ev_forecast_0_trial_{}.csv".format(type, trial)))
```

```python
myData = {"Individual ({})".format(label) : arr1, 
          "Joint ({})".format(label) : arr3, 
          "Hybrid ({})".format(label) : arr4, "Optimal Hybrid (PF)" : arr5}
```

```python
fig = plt.figure(figsize = (8, 5))
plt.boxplot(myData.values(), labels=myData.keys())
plt.ylabel("Cost [$]", fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
# plt.savefig("boxplot_cases_MPC.png")
plt.show()
```

```python
path  = "{}/Code/Results/case_4/cost_list/".format(os.getcwd())
path1 = "{}/Code/Results/case_1/cost_list/".format(os.getcwd())
path2 = "{}/Code/Results/case_3/cost_list/".format(os.getcwd())
```

```python
arr_00  = np.array([])
arr_25  = np.array([])
arr_50  = np.array([])
arr_75  = np.array([])
arr_100 = np.array([]) 

for trial in range(1, 51):
    arr_00  = np.append(arr_00, np.loadtxt(path1 + "cost_list_{}_1_ev_forecast_0_trial_{}.csv".format("determ", trial)))
    arr_25  = np.append(arr_25, np.loadtxt(path + "cost_list_{}_4_ev_forecast_0_25_trial_{}.csv".format("determ", trial)))
    arr_50  = np.append(arr_50, np.loadtxt(path + "cost_list_{}_4_ev_forecast_0_trial_{}.csv".format("determ", trial)))
    arr_75  = np.append(arr_75, np.loadtxt(path + "cost_list_{}_4_ev_forecast_0_75_trial_{}.csv".format("determ", trial)))
    arr_100 = np.append(arr_100, np.loadtxt(path2 + "cost_list_{}_3_ev_forecast_0_trial_{}.csv".format("determ", trial)))
```

```python
label = "PF"

myData = {"0% ({})".format(label)  : arr_00,
          "25% ({})".format(label)  : arr_25, 
          "50% ({})".format(label)  : arr_50, 
          "75% ({})".format(label)  : arr_75,
          "100% ({})".format(label) : arr_100}
```

```python
fig = plt.figure(figsize = (8, 5))
plt.boxplot(myData.values(), labels=myData.keys(), medianprops = dict(color = "#AA3377", linewidth = 1.5))
plt.ylabel("Cost [$]", fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("Percent Shared", fontsize = 15)
#plt.savefig("boxplot_hybrid_PF.png")
plt.show()
```
